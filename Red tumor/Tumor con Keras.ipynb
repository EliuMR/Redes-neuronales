```
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.core import Dense
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
data=pd.read_csv('/content/drive/MyDrive/archive.zip (Unzipped Files)/data.csv')
data.head()
```

```
Ys=np.array(data['diagnosis'])
n=len(Ys)
Y=np.zeros(n)
for i in range(n):
  if Ys[i]=='M':
    Y[i]=0
  else:
    Y[i]=1
print(len(Y))
X=np.zeros((30,569))
Nombre=['radius','texture','perimeter','area','symmetry','fractal_dimension','concavity','concave points','compactness','smoothness']
Sufijo=['_mean','_se','_worst']
m=0
for i in Sufijo:
  for j in Nombre:
    X[m]=np.array(data[j+i])
    m=m+1
X=X.T
Train_X,Test_X,Train_Y,Test_Y=train_test_split(X,Y,test_size=.2)
print(Train_X.shape,Train_Y.shape)
print(Test_X.shape,Test_Y.shape)
```

```
model = Sequential()
model.add(Dense(16, input_dim=30, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
model.fit(Train_X, Train_Y, epochs=40)
```

```
scores = model.evaluate(Test_X,Test_Y)
print('La evaluación de modelo es: ',scores[1]*100)
```
